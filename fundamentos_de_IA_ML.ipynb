```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("compra.csv")
df.head()
```

```python
print ("linhas, colunas ", df.shape) #retorna as n linhas, m colunas = (n, m)
print (df["Compra"].value_counts()) #mostra quantidade das diferentes instancias de uma feature
print (df["Compra"].value_counts(normalize=True)) #mostra agora com a normalizacao ente 0 e 1
print(df["Altura_m"].values)
print(df["Altura_m"].size)
```

```python
def Tendencias(df, nomeAtributo):
  #media
  soma = 0
  atributo = df[nomeAtributo]
  tamanho = len(atributo)
  for i in range(tamanho):
    soma = soma + atributo[i]

  media = soma / 150
  mediana = atributo.median()

  moda = atributo.mode().iloc[0]

  print(f"--- Tendências para: {nomeAtributo} ---")
  print(f"Média: {media:.2f}")
  print(f"Mediana: {mediana:.2f}")
  print(f"Moda: {moda:.2f}")

  #moda

Tendencias(df, "Altura_m")
```

```python
def DispersaoEPosicao(df, nomeAtributo):
    atributo = df[nomeAtributo]


    minimo = atributo.min()
    maximo = atributo.max()
    amplitude = maximo - minimo


    # Q1 é os 25%, Q2 é os 50% (que é igual a mediana), Q3 é os 75%
    q1 = atributo.quantile(0.25)
    q2 = atributo.quantile(0.50) # = atributo.median()
    q3 = atributo.quantile(0.75)
    iqr = q3 - q1

    # --- Variância e Desvio-Padrão ---

    variancia = atributo.var()
    desvio_padrao = atributo.std()

    print(f"--- Análise para {nomeAtributo} ---")
    print(f"Mínimo: {minimo:.2f} | Máximo: {maximo:.2f} | Amplitude: {amplitude:.2f}")
    print(f"Q1 (25%): {q1:.2f} | Q2/Mediana (50%): {q2:.2f} | Q3 (75%): {q3:.2f}")
    print(f"IQR: {iqr:.2f}")
    print(f"Variância: {variancia:.4f} | Desvio-Padrão: {desvio_padrao:.4f}")

DispersaoEPosicao(df, "Altura_m")
```

```python
#Idade	Altura_m	Salario_R$	Filhos
def Tabela(df):
  print(df.describe())
  DispersaoEPosicao(df, "Altura_m")
  DispersaoEPosicao(df, "Idade")
  DispersaoEPosicao(df, "Salario_R$")
  DispersaoEPosicao(df, "Filhos")

Tabela(df)
```

```python
df_numerico = df.select_dtypes(include=['number'])

#Calculamos a matriz de correlação (o padrão já é 'pearson')
matriz_correlacao = df_numerico.corr(method='pearson')

#Configuramos o tamanho do gráfico para ficar bonito
plt.figure(figsize=(8, 6))

# Criamos o Heatmap
# annot=True: Coloca os números dentro dos quadradinhos
# cmap='coolwarm': Usa cores de azul (frio/negativo) até vermelho (quente/positivo)
# vmin e vmax: Força a escala de cores ir exatamente de -1 a 1
# fmt=".2f": Deixa os números com 2 casas decimais
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt=".2f", linewidths=0.5)

plt.title('Matriz de Correlação de Pearson')
plt.show()
```
